## Good Practices

Clarity and reproducibility matters just as much as correctness.

Use Virtual Environments

Good research code should make it easy to answer questions such as:

- Which input data was used?
- Which parameters were chosen?
- Which environment was the code run in?
- Can someone else rerun this later?

How can we make these questions easily answerable?

Small habits already help a lot:

- keep scripts focused,
- use meaningful names (Functions should describe actions),
- avoid hidden manual steps,
- write down assumptions near the code that depends on them.

Also, stick to the PEP 8 Python style guide. The most important beginner takeaway is simple: prefer readable code over clever code.

That usually means:

- Use pathlib for file Paths
- Keep Imports Organized: only import what you need, import everything at the beginning/one point
- short functions. Break large functions into smaller ones,
- descriptive variable names,
- consistent indentation,
- spaces around operators,
- blank lines to separate ideas,
- Start variables with lower case,
- Avoid repetitions (use functions if code doubles)
- separate words of a variable with underscores. E.g., this_variable_stores_int = 10,
- Pythonic convention is to use ALL_UPPER_CASE to define constants: THIS_IS_A_CONSTANT = 10 (can be changed later, but should not be changed, since constant)

```python
# Example for code that is difficult to read and understand. 
# The function name does not describe its action
# The code is not commented and the variable names are not descriptive.

def lets_see(x):
    new = 0
    for i in range(len(x)):
        if x[i] == 10:
            new += 1
            print("Oh no!")
        elif x[i] > 10:
            print("up")
        elif x[i] < 10:
            print("its low")
    return new
```

## Question: How to make this more readable?

```python
def count_errors(cont_measurement: list) -> int:
    """Takes a list of continous measurements and returns the number of errors"""
    tot_num_errors = 0
    for inv_measure in cont_measurement:
        if inv_measure == 10: # we assume that every time we measure a 10, its an error
            tot_num_errors += 1
            print("An error occured in the measurement")
        elif inv_measure > 10:
            print("No error, high values detected")
        elif inv_measure < 10:
            print("No error, low values detected")
    return tot_num_errors
```

```python
def return_integers(inv_list: list) -> list: 
    '''
    This function takes a list as input and returns a new list containing only the integer elements from the original list.
    '''
    #create empty list to store integers
    int_list = [] 
    
    # loop over all entries of input list and save entry in new list, if it is an integer
    for entry in inv_list:
        if isinstance(entry, int):
           int_list.append(entry)
        else:
            continue
        
    # return the new list with all integers
    return int_list
```
