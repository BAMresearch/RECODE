```python
# We defined a function to search Path directories for specific files
def find_file(directory, filename):
    """
    This function takes the directory input and tries to find the
    specified filename inside of it in a recursive way using rglob()
    from pathlib.
    """
    for path in directory.rglob(filename):
        return path
    print(f"I couldn't find the {filename} in {directory}")
```

```python
import pandas as pd
import numpy as np
from pathlib import Path
import os
import matplotlib.pyplot as plt

# With extracting beforehand
home_dir = Path.home()

spectra_folder = home_dir / "Miniconda3/envs/jlab/Scripts/recode/synth_spectra/"

# read a single spectrum with pandas
spectrum_0 = pd.read_csv(spectra_folder / "spectrum_0.csv")
```

```python
spectrum_0
```

```python
# we now want to read all spectra and create a single pandas dataframe from them

# to iterate over the individual .csv-files, we need a list with all their names
# We can either create one by hand, or use the os package to generate them automatically:
spectra_in_folder = os.listdir(spectra_folder)
spectra_in_folder
```

```python
# create a list with dataframes, which are put together after the loop
spectra = []

for spectrum in spectra_in_folder:
    if spectrum.endswith(".csv"):
        selected_spectrum = pd.read_csv(spectra_folder / spectrum)
        spectra.append(selected_spectrum)
        
```

```python
df_v1 = pd.concat(spectra, ignore_index=True)
df_v1.head()
```

```python
df_v1.to_csv(spectra_folder / "combined_spectra.csv", index=False)
```

```python
# concatenate each new dataframe to the existing dataframe
spectra = []

for idx_spec in range(len(spectra_in_folder)):
    if spectra_in_folder[idx_spec].endswith(".csv"):
        if idx_spec == 0:
            df_v2 = pd.read_csv(spectra_folder / spectra_in_folder[idx_spec])
        else:
            spec_to_add = pd.read_csv(spectra_folder / spectra_in_folder[idx_spec])
            df_v2 = pd.concat([df_v2,spec_to_add], ignore_index=True)
        
```

```python
df_v2.head()
```

```python
df_v2_copy = df_v2.copy()
```

```python
df_v2_copy.shape
```

```python
# changing individual entries
df_v2_copy.iloc[0,0] = 0

# This takes too long for all entries...
```

```python
# we can do it row by row

for spectrum in range(df_v2_copy.shape[0]):
    cols_below_zero = df_v2_copy.iloc[spectrum,:]<0
    df_v2_copy.iloc[spectrum, cols_below_zero] = 0
```

```python
# column by column

# reset the copy first, otherwise, everything is set to 0 already
df_v2_copy = df_v2.copy()

for spectrum in range(df_v2_copy.shape[1]):
    rows_below_zero = df_v2_copy.iloc[:,spectrum]<0
    df_v2_copy.iloc[rows_below_zero, spectrum] = 0
```

```python
# all together (the simples option)
df_v2_copy = df_v2.copy()

df_v2_copy[df_v2_copy<0] = 0

# df_v2_copy<0 -> this statement returns a boolean that is true for all entries below 0, and false for all entries above 0
# the boolean has the same shape as our orinigal dataframe. If we use the [] to index the entries in our dataframe, every
# entrie that is true will be set to the value after the = sign. Every entry that is false will remain the same. 
```
