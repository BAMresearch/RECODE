Exercise 2 - Define a function to find a filename using rglob

Explanation: in the cells before, we copy-pasted the code 2 times to define the variables `example_file` and `test_file`. When this happens, it is a good programming practice to combine such logic in a single function, and then later simply call for the function instead of copy-pasting all the lines of code all the time.

```python
def find_file(directory, filename):
    """
    This function takes the directory input and tries to find the
    specified filename inside of it in a recursive way using rglob()
    from pathlib.
    """
    for path in directory.rglob(filename):
        return path
    print(f"I couldn't find the {filename} in {directory}")
```

```python
example_file = find_file(home_dir, "example_data.dat")
test_file = find_file(home_dir, "test.dat")
print(example_file, test_file)
```

```python
# Some operations with strings
my_example_str = "# This is a header\n## This is a subheader\n### A three-level-down header"

print(my_example_str.split("\n"))  # `split()` breaks the string and convert it into a list
print(my_example_str.startswith("#"), my_example_str.startswith("##"))  # startswith() check if the string starts with the character
print(my_example_str.replace("#", "$"))  # `replace()` changes the first character for the second
```

```python
data = []
for row in lines:
    if row.startswith("#"):
        continue
    else:
        if "NaN" in row:
            continue
        else:
            new_row = row.replace("\n", "")
            x, y = new_row.split(" ")
            # print("This is x: ", x, type(x), "\tThis is y:", y, type(y))
            x_now_as_a_float = float(x)  # in Java: xNowAsAFloat
            y_now_as_a_float = float(y)
            data.append([x_now_as_a_float, y_now_as_a_float])
print(data)
```
