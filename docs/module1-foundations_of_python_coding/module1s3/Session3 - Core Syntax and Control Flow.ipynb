## Conditionals

```python
temperature = 20

if temperature < 20:
    print("It's cold")
elif temperature > 20:
    print("It's warm")
elif temperature == 20:
    print("It's moderate")
else:  # other cases, which are none
    print("This should never happen")
```

## List Comprehensions

```python
# [fruit.upper()                 # sth. to do with each item
#     for fruit in fruits        # iterate over each item
#     if fruit.startswith("a")]  # optional condition, like a filter
new_fruits = [fruit.upper() for fruit in fruits if fruit.startswith("a")]
new_fruits
```

```python
new_fruits = []
for fruit in fruits:
    if fruit.startswith("a"):
        new_fruits.append(fruit.upper())
new_fruits
```

```python
[n**3 for n in range(5)]
```

### Common pitfall of lists and strings

```python
def print_items(items):
    for item in items:
        print(f"{item=}")

# Intended usage:
print_items(fruits)
```

```python
print_items("cherry")
```

```python
# fixed print_items function to handle both strings and lists of strings
def print_items(items):
    if isinstance(items, str):  # check if items is a string
        items = [items]  # convert it to a list with one item
    for item in items:
        print(f"{item=}")

print_items("cherry")
```
