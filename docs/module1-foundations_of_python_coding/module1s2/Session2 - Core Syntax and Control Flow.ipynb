### Exercise: write a function to transform from Celsius to Kelvin

```python
def to_kelvin(temperature_in_celsius):
    # This function converts a temperature from Celsius to Kelvin
    temperature_in_kelvin = temperature_in_celsius + 273.15
    return temperature_in_kelvin

to_kelvin(25)
```

```python
print("25 degrees celsius in kelvin is", to_kelvin(25))
```
