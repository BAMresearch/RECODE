```python
number_list = [1, 2, 3, 202, 344, 33, 90, 2026]
for element in number_list:
    if element % 2 == 0:  # even number
        print("Even number: ", element)
    else:
        print("Odd number: ", element)
```

## Dictionaries and nesting

```python
example_dict = {
    "key1": 10,
    "key2": "another string",
    "key3": 3 + 3j,
    "key4": [1, 2, 3],
    "key5": {
        "inside_key1": 22,
        "inside_key2": 39.0,
        "inside_key3": {
            "inside_inside_key1": "temperature"
        }
    }
}
```

```python
for key, value in example_dict.items():
    print(key, value)
```

```python
for key, value in example_dict.items():
    if isinstance(value, dict):
        for subkey, subvalue in value.items():  # this can be nested as many times as nested dicts are
            print(subkey, subvalue)
    else:
        print(key, value)
```
